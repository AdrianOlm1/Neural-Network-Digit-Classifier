```python
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
train_data = datasets.MNIST(
    root  = 'data',
    train = True,
    transform = ToTensor(),
    download = True,
)
tst_data = datasets.MNIST(
    root  = 'data',
    train = False,
    transform = ToTensor(),
    download = True,
)
```

```python
train_data.data.shape
```

```python
tst_data.data.shape
```

```python
loaders  = {
    'train': DataLoader(train_data,
                        batch_size = 100,
                        shuffle = True,
                        num_workers = 1),
    'test': DataLoader(tst_data,
                        batch_size = 100,
                        shuffle = True,
                        num_workers = 1),
}
```

```python
loaders 
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64 * 7 * 7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = torch.max_pool2d(x, 2, 2)
        x = torch.relu(self.conv2(x))
        x = torch.max_pool2d(x, 2, 2)
        x = x.view(-1, 64 * 7 * 7)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model = CNN().to(device)

optimizer = optim.Adam(model.parameters(), lr=0.001)

loss_fn = nn.CrossEntropyLoss()

def train(epoch):
    model.train()
    for batch_index, (data, target) in enumerate(loaders['train']):
        data, target = data.to(device), target.to(device)
        optimizer.zero_grad()
        output = model(data)
        loss = loss_fn(output, target)
        loss.backward()
        optimizer.step()
        if batch_index % 20 == 0:
            print(f'Train Epoch: {epoch} [{batch_index * len(data)}/{len(loaders["train"].dataset)} ({100. * batch_index / len(loaders["train"]):.0f}%)]\t{loss.item():.6f}')

def test():
    model.eval()

    test_loss = 0
    correct = 0

    with torch.no_grad():
        for data, target in loaders['test']:
            data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += loss_fn(output, target).item()
            pred = output.argmax(dim = 1, keepdim=True)
            correct += pred.eq(target.view_as(pred)).sum().item()

    test_loss /= len(loaders['test'].dataset)
    print(f'\nTest set: Average loss: {test_loss:.4f}, Accuracy {correct}/{len(loaders["test"].dataset)} ({100. * correct / len(loaders["test"].dataset):.0f})')
```

```python
for epoch in range(10):
    train(epoch)
    test()
    
torch.save(model.state_dict(), "sudoku_cnn_weights.pth")
torch.save(model, "sudoku_cnn_full.pth")
print("Model saved successfully!")
```

```python
import cv2
import numpy as np
import torch
import matplotlib.pyplot as plt

# Load your trained model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")  # Ensure correct device
model.to(device) 
model.eval()

# Open the webcam
cap = cv2.VideoCapture(0)  # 0 is the default webcam, change if needed

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Invert colors (MNIST is black on white, but the camera may capture white on black)
    gray = cv2.bitwise_not(gray)

    # Resize to match MNIST (28x28)
    resized = cv2.resize(gray, (28, 28))

    # Normalize to match the training data (MNIST is in range 0-1)
    normalized = resized.astype(np.float32) / 255.0

    # Convert to tensor (1,1,28,28) for CNN input
    tensor_input = torch.tensor(normalized, dtype=torch.float32).unsqueeze(0).unsqueeze(0).to(device)

    # Run prediction
    with torch.no_grad():
        output = model(tensor_input)
        prediction = output.argmax(dim=1).item()

    # Display the prediction on the camera feed
    cv2.putText(frame, f'Prediction: {prediction}', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show the original camera feed
    cv2.imshow('Camera Feed', frame)

    # Show the preprocessed image (grayscale input)
    resized_display = cv2.resize(resized, (200, 200), interpolation=cv2.INTER_NEAREST)
    cv2.imshow('Processed Input (28x28)', resized_display)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader
from PIL import Image
from BACKTRACKER import solveSudoku
############################################################################################################
###################### Preprocesses image for the NN
############################################################################################################

transform = transforms.Compose([
    transforms.Resize((28, 28)),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

def resize_and_pad(cell, target_size=28):
    """Resizes while keeping aspect ratio and adding padding if necessary."""
    h, w = cell.shape

    if h == w:  # If already square, just resize
        return cv2.resize(cell, (target_size, target_size))
    
    scale = target_size / max(h, w)
    resized = cv2.resize(cell, (int(w * scale), int(h * scale)))

    # Create a blank 28x28 image
    padded = np.zeros((target_size, target_size), dtype=np.uint8)
    pad_x = (target_size - resized.shape[1]) // 2
    pad_y = (target_size - resized.shape[0]) // 2

    # Paste resized digit onto center of blank image
    padded[pad_y:pad_y+resized.shape[0], pad_x:pad_x+resized.shape[1]] = resized

    return padded

def preprocess(cell):
    """Preprocess Sudoku cell for CNN digit recognition."""
    
    # Resize & remove noise
    cell = cv2.GaussianBlur(cell, (3, 3), 0)

    # Apply adaptive thresholding (Otsu’s method)
    _, cell = cv2.threshold(cell, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Dilate to prevent erasing parts of digits
    kernel = np.ones((2,2), np.uint8)
    cell = cv2.dilate(cell, kernel, iterations=1)

    # Resize with padding to maintain aspect ratio
    cell = resize_and_pad(cell, 28)

    # Convert to PyTorch tensor & normalize
    cell = Image.fromarray(cell)
    cell = transforms.ToTensor()(cell)
    cell = transforms.Normalize((0.1307,), (0.3081,))(cell)
    cell = cell.unsqueeze(0)  # Add batch dimension
    
    return cell



############################################################################################################
###################### Turns image into grayscale and then performs adapative threshold on image
############################################################################################################

image = cv2.imread('test.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(image)
plt.title("Gray scaled image")
plt.show()

blurred = cv2.GaussianBlur(gray, (5,5), 0)

thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11, 2)

plt.imshow(thresh, cmap="gray")
plt.title("Threshold")
plt.show()

############################################################################################################
###################### Detects a sudoku grid
############################################################################################################

contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contours = sorted(contours,key= cv2.contourArea, reverse=True)

sudoku_contour = None
for contour in contours:
    epsilon = 0.02 * cv2.arcLength(contour, True)
    aprox = cv2.approxPolyDP(contour, epsilon, True)
    if len(aprox) == 4:
        sudoku_contour = aprox
        break

grid_image = image.copy()
cv2.drawContours(grid_image, [sudoku_contour], -1, (0,255,0), 3)

plt.imshow(cv2.cvtColor(grid_image, cv2.COLOR_BGR2RGB))
plt.title("Located Grid")
plt.show()

############################################################################################################
###################### Warps the photo to align the grid nicely
############################################################################################################

sudoku_contour = sudoku_contour.reshape(4, 2)

sudoku_contour = sorted(sudoku_contour, key=lambda x: (x[1], x[0]))
top_left, top_right = sorted(sudoku_contour[:2], key=lambda x: x[0])
bottom_left, bottom_right = sorted(sudoku_contour[2:], key=lambda x: x[0])

size = 450
pt1 = np.array([top_left, top_right, bottom_right, bottom_left], dtype="float32")
pt2 = np.array([[0,0], [size, 0], [size, size], [0, size]], dtype="float32")

matrix = cv2.getPerspectiveTransform(pt1, pt2)
warped = cv2.warpPerspective(gray, matrix ,(size,size))

plt.imshow(warped, cmap="gray")
plt.title("warped sudoku grid")
plt.show()

############################################################################################################
###################### Turns whole photo into cells to be preprocessed
############################################################################################################

cell_size = size // 9
cells = []

for i in range(9):
    row = []
    for j in range(9):
        x_start, y_start = j * cell_size, i*cell_size
        x_end, y_end = (j+1) *cell_size, (i+1) *cell_size

        cell = warped[y_start:y_end, x_start:x_end]
        row.append(cell)

    cells.append(row)

plt.imshow(cells[0][0], cmap="gray")
plt.title("Example of grid taken")
plt.show()

############################################################################################################
###################### Creates an array from the grid of photos and cnn model
############################################################################################################

fig, axes = plt.subplots(9, 9, figsize=(8, 8))

for i in range(9):
    for j in range(9):
        preprocessed_cell = preprocess(cells[i][j])  # Apply preprocessing
        img = preprocessed_cell.squeeze().numpy()  # Convert tensor to NumPy array

        ax = axes[i, j]
        ax.imshow(img, cmap="gray")  # Show preprocessed cell
        ax.axis("off")

plt.suptitle("Preprocessed Sudoku Grid Cells", fontsize=16)
plt.show()

model = CNN()
model.load_state_dict(torch.load("sudoku_cnn_weights.pth", map_location=torch.device("cpu")))

model.eval()  # Set to evaluation mode

print("Model loaded successfully!")

sudoku_grid = np.zeros((9,9), dtype=int)

for i in range(9):
    for j in range(9):
        cell = preprocess(cells[i][j])

        with torch.no_grad():
            output = model(cell)
            digit = torch.argmax(output).item()
            confidence = torch.softmax(output, dim=1).max().item()
        
        print(f"Predicted: {digit}, Confidence: {confidence:.2f}")
        
        if confidence < 0.99:  # Check probability of the predicted digit
            digit = 0

        sudoku_grid[i][j] = digit

for row in sudoku_grid:
    print(" ".join(map(str, row)))
solveSudoku(sudoku_grid)
print("\nSolved Sudoku Grid:")
for row in sudoku_grid:
    print(" ".join(map(str, row)))
```
